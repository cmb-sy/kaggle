# 概要
エクアドルにある Favorita 店舗で販売されている何千もの製品の売上を予測する。
訓練データには、日付、店舗と製品の情報、その商品がプロモーションされているか、販売数が含まれる。
追加ファイルには、モデルの構築に役立つ補足情報がある。

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Jupyter Notebook環境でのマジックコマン.グラフやプロットがNotebook上で直接表示。
%matplotlib inline
import warnings
# プログラムが実行される際に発生する警告メッセージを無視
warnings.filterwarnings('ignore')
```

# データの読み込み

```python
df_train = pd.read_csv('data/train.csv',usecols=['store_nbr', 'family', 'date', 'sales', 'onpromotion'],
                       dtype={'store_nbr': 'category', 'family': 'category', 'sales': 'float64',},parse_dates=['date'],infer_datetime_format=True)
df_test = pd.read_csv('data/test.csv')
df_stores = pd.read_csv('data/stores.csv')
df_oil = pd.read_csv('data/oil.csv')
df_holidays = pd.read_csv('data/holidays_events.csv')
df_transactions = pd.read_csv('data/transactions.csv')
```

# データの分析と加工

```python
print("Train Data:")
print(df_train.info())
print("\nTest Data:")
print(df_test.info())
print("\nStores Data:")
print(df_stores.info())
print("\nOil Data:")
print(df_oil.info())
print("\nHolidays Data:")
print(df_holidays.info())
print("\nTransactions Data:")
print(df_transactions.info())
```

データの結合とidの削除

```python
df = pd.concat([df_train, df_test])
df.drop(columns='id', inplace=True) # inplace=Trueで元のdfを上書きしてしまう。

print(df.shape)
print(df.columns)
```

結合後のデータを確認する。

```python
print("統計データ", df.describe())
print()
print("nullの確認", df.isnull().sum())
print()
print("概要", df.info())
```

dateデータをdatetimeオブジェクトへ変換。
カテゴリカルデータのユニークな値数を取得。

```python
# 'date' 列をdatetimeオブジェクトへ変換。（日時を扱いやすくする）
df['date'] = pd.to_datetime(df['date'])

# カテゴリデータのユニーク数を表示
for col in ['store_nbr', 'family']:
    print(f"Unique values in {col}: {df[col].nunique()}")
```

日付データの分解

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['week'] = df['date'].dt.day_of_week
# isinでTrueかFalseを返し、astypeで0,1へ変換する。1が土日に該当する。
df['weekend'] = df['week'].isin([5, 6]).astype(int)
```

```python
df.describe().T
print('null数')
df.isnull().sum()
```

## データの可視化
### date vs sales
### onpromotion vs sales

```python
# データの可視化
plt.figure(figsize=(8, 4))
plt.plot(df['date'], df['sales'], label='Sales')
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Explore relationships between features
plt.figure(figsize=(8, 4))
plt.scatter(df['onpromotion'], df['sales'], alpha=0.5)
plt.title('Sales vs. On Promotion')
plt.xlabel('On Promotion')
plt.ylabel('Sales')
plt.show()
```

## 月毎の平均の売上

```python
plt.figure(figsize=(8, 4))
df.groupby('month')['sales'].mean().plot(kind='bar', color='skyblue') # dfを月ごとにグループ化し、各月の売上 (sales 列) の平均を計算。
plt.title('Average Sales by Month')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.show()
```

## 休日と平日の売上
df['date'].isin(df_holidays['date'])：df['date'] 列内の各要素が df_holidays['date'] 列に含まれているかどうかを確認
.astype(int)：ブール型のシリーズを整数型に変換します。True は1に、False は0に変換。

```python
# 'date'列がdf_holidays DataFrameの'date'列に一致する場合は1（休日）、一致しない場合は0（非休日）になります。
df['is_holiday'] = df['date'].isin(df_holidays['date']).astype(int)

df.groupby('is_holiday')['sales'].mean().plot(kind='barh')
plt.title('Average Sales on Holidays vs. Non-Holidays')
plt.yticks([0, 1], ['Non-Holiday', 'Holiday'])
plt.ylabel('Holiday')
plt.xlabel('Average Sales')
plt.show()
```
